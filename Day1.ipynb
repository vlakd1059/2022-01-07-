### Requests 사용법

```python
# 서버에 페이지 정보를 요청할때 사용하는 라이브러리
import requests as req
```

### BeautifulSoup 사용하기

```python
from bs4 import BeautifulSoup as bs
```

```python
# 파싱 bs(파싱할 데이터, 파싱방법)
soup=bs(res.text,'lxml')
soup
```

```python
soup.select_one('a.nav')
```

```python
soup.select('a.nav')[2]
```

```python
soup.select('a.nav')[2].text
```

### 멜론차트

```python
h={'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/96.0.4664.110 Safari/537.36'}
```

```python
url='https://www.melon.com/chart/index.htm'
res=req.get(url,headers=h)
res
#<Response [406]> : 응답실패
```

```python
res.text
```

```python
soup=bs(res.text,'lxml')
soup
```

```python
title=soup.select('div.ellipsis.rank01>span>a')
```

```python
for i in title:
    print(i.text)
```

```python
singer=soup.select('div.ellipsis.rank02>span>a')
```

```python
for i in singer:
    print(i.text)
```

```python
singer=soup.select('div.ellipsis.rank02>a:nth-child(1)')
```

```python
for i in singer:
    print(i.text)
```

```python
title_list=[]
singer_list=[]
rank_list=[]

for i in range(len(title)):
    title_list.append(title[i].text)
    singer_list.append(singer[i].text)
    rank_list.append(i+1)
```

```python
import pandas as pd
```

```python
dic={'노래제목' : title_list, '가수' : singer_list, '순위':rank_list}
```

```python
df=pd.DataFrame(dic)
```

```python
df
```

```python
df=df.set_index('순위')
```
